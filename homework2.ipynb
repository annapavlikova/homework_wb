Необходимо написать с использованием numpy реализацию одного из следующих алгоритмов: 

- heapsort (вариант: 1,5,9,13,17,21)

- mergesort (вариант: 2,6,10,14,18,22)

- stable (вариант: 3,7,11,15,19)

- quicksort (вариант: 4,8,12,16,20) 

Сравнить скорость выполнения этих алгоритмов на чистом python, вашей реализации и numpy.sort() 

```python
#quicksort с использованием numpy
def quicksort(x):
    if x.size == 1 or x.size == 0:
        return x
    else:
        pivot = x[0]
        i = 0
        for j in range(x.size-1):
            if x[j+1] < pivot:
                x[j+1],x[i+1] = x[i+1], x[j+1]
                i += 1
        x[0],x[i] = x[i],x[0]
        first_part = quicksort(x[:i])
        second_part = quicksort(x[i+1:])
        #print(first_part)
        #print(second_part)
        first_part = np.append(first_part, x[i])
        return (np.concatenate([first_part, second_part]))

g=np.array([84,32,94,72,93,62,94])
print(quicksort(g))

a = np.array([random.randint(0, 1000) for i in range(1000)])
```

```python
#время qiucksort с использованием numpy
%%timeit
quicksort(a)
```

```python
#время np.sort
%%timeit
np.sort(a)
```
