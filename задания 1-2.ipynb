**Introduction**

1. Может ли быть в таблице реляционной базы данных несколько строк с повторяющимся PK? Почему?

В реляционной базе не может быть строк с одинаковым ключом, так как каждая запись должна однозначно определяться значением ключа. Если будут повторяющиеся значения, однозначная идентификация станет невозможной.

2. К какому типу относится субд Redis? В чем особенность данной субд при работе с памятью?

Redis - это нереляционная резидентная СУБД с открытым исходным кодом, работающая со структурами данных типа «ключ — значение».

Особенности при работе с данными:
Redis — это база данных, размещаемая в памяти, которая используется, в основном, в роли кеша, находящегося перед другой базой данных. Кеш, основанный на Redis, помогает улучшить производительность приложений, и смягчает нагрузку центральной базы данных приложения. Клиент обращается к приложению, а оно получает необходимые для выполнения его запроса данные. Сначал приложение обращается к кешу Redis представленному главной базой данных (Main). Если данные в кеше есть, произошло попадание кеша, выполняется обычный возврат данных. Если произошёл промах кеша, система обращается к постоянному хранилищу. Данные из него  загружаются в кеш, после чего ими сможет воспользоваться приложение. Традиционные СУБД используют жёсткие диски (HDD) или твердотельные накопители (SSD), а Redis - читает данные из памяти.


3. Что является аналогом таблицы (в реляционной СУБД) в MongoDB? Что является аналогом строки/записи(в реляционной СУБД) в MongoDB? Приведите небольшой пример того, как может выглядеть данный аналог.
    
    В MongoDB аналог таблиц - коллекции, аналог строк - документы.
    Документ - набор пар ключ-значение, например:
    
    {
    "name": "Anna",
    
    "surname": "Pavlikova",
    
    "age": "21"
    }
    
    В коллекции может быть несколько документов.

4. Выберите наиболее подходящие СУБД для следующих задач:
    
    - Анализ поведения знаменитостей в социальной сети.
    для социальных сетей хорошо подходят графовые СУБД, 
    поэтому можно использовать, например, Neo4j, Amazon Neptune, InfiniteGraph, InfoGrid
    
    - Онлайн-аналитика действий пользователей на сайте онлайн магазина. Включая информацию о клиентах, товарах, заказах.

    Например, Oracle, MySQL, Microsoft SQL Server, PostgreSQL.
    
    - Анализ документации по фильмам и сериалам, включая: название, год выпуска, список актеров, жанр и т.д.
    Можно использовать реляционные или документные, так как в таблицах хранят объекты с разной структурой. Примеры - CouchDB, MongoDB.
    
    - Проектирование наиболее оптимальных маршрутов поставок для маркетплейса

    - Хранение кэшированных HTML страниц пользователей, посещающих сайт
    Для кэширования данных выгодно использовать СУБД типа ключ-значение. Они позволяют работать полностью в памяти и задавать срок жизни записи, удаляя устаревшие данные. Примеры таких СУБД - Redis и Memcached.
